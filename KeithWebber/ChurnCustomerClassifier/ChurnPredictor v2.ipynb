```python
churn = df[churn_fields]
churn.info()
```

```python
def determine_churn(row):
    if np.nan_to_num(row["Churn_Notes__c"]) != 0:
        return 1
    elif np.nan_to_num(row["GAPPS_Churn_Date__c"]) != 0:
        return 1
    elif np.nan_to_num(row["GAPPS_Churn_Reason__c"]) != 0:
        return 1
    elif np.nan_to_num(row["GAPPS_Churn__c"]) == True:
        return 1
    elif np.nan_to_num(row["SFDC_Churn__c"]) == True:
        return 1    
    else:
        return 0
    
    
df["churned"] = 0

for index, row in churn.iterrows():
    df["churned"][index] = determine_churn(row)

df["churned"]
```
